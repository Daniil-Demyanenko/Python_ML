### Подготовка данных

Нормализуем датасет, поскольку это важно для большинства алгоритмов классификации

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = df.drop(columns=["Class", "index"])
df_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
df_scaled["Class"] = df["Class"]

df_scaled.head()
```

Создадим разные наборы данных с помощью разных методов балансировки датасета.

#### Under-sampling

Исключим из датасета часть законных транзакций, сбалансировав его. Таким образом мы теряем большую часть данных о легальных транзакциях, но повышаем представленность класса нелегальных.

```python
illegal_count = len(df_scaled[df_scaled["Class"] == 1])

legal_sample = df_scaled[df_scaled["Class"] == 0].sample(n=illegal_count, random_state=42)

df_undersampled = pd.concat([legal_sample, df_scaled[df_scaled["Class"] == 1]], axis=0)

df_undersampled["Class"].value_counts()
```

#### Дублирование (Repeat Sampling)
Этот метод просто дублирует образцы из меньшего класса, чтобы повысить представленность образцов меньшего класса.  
Я предлагаю перед дублированием разбить данные на тренировочную и тестовую выборки, чтоб не тестировать модель на дубликатах данных, которые были в обучающей выборке. 

```python
df_rs_test = pd.concat([
    df_scaled[df_scaled["Class"] == 1][:100],
    df_scaled[df_scaled["Class"] == 0][:100]])

df_rs_train = pd.concat([
    df_scaled[df_scaled["Class"] == 1][100:],
    df_scaled[df_scaled["Class"] == 0][100:]])

# вычислим, сколько раз нужно продублировать 
count_illegal = len(df_rs_train[df_rs_train["Class"] == 1])
count_llegal = len(df_rs_train[df_rs_train["Class"] == 0])
c = count_llegal - count_illegal

duplicated_rows = df_rs_train[df_rs_train['Class'] == 1].sample(n=c, replace=True, random_state=42)
df_rs_train = pd.concat([df_rs_train, duplicated_rows])

df_rs_train["Class"].value_counts()
```

#### Синтетическое увеличение (SMOTE - Synthetic Minority Over-sampling Technique)
SMOTE создает новые образцы для меньшего класса, используя интерполяцию между соседними образцами. Это позволяет увеличить разнообразие данных и уменьшить переобучение.

```python
from imblearn.over_sampling import SMOTE

X = df_scaled.drop(columns=["Class"])
y = df_scaled["Class"]

smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

df_SMOTE = pd.DataFrame(X_resampled, columns=X.columns)
df_SMOTE["Class"] = y_resampled

df_SMOTE["Class"].value_counts()
```

#### ADASYN (Adaptive Synthetic Oversampling)
Это метод, который помогает улучшить баланс классов, создавая синтетические примеры из миноритарного класса. В отличие от SMOTE, который генерирует одинаковое количество синтетических образцов для всех миноритарных примеров, ADASYN адаптивно создает разное количество синтетических образцов в зависимости от оценки локального распределения класса, который нужно увеличить.

```python
from imblearn.over_sampling import ADASYN

adasyn = ADASYN(sampling_strategy='auto', random_state=42)

X_resampled, y_resampled = adasyn.fit_resample(X, y)

df_ADASYN = pd.DataFrame(X_resampled, columns=X.columns)
df_ADASYN["Class"] = y_resampled

df_ADASYN["Class"].value_counts()
```

#### Tomek Links
Это метод оверсемплинга, основанный на удалении так называемых “связей Томека” между классами.

```python
from imblearn.under_sampling import TomekLinks

tomek = TomekLinks()

X_resampled, y_resampled = tomek.fit_resample(X, y)

df_Tomek = pd.DataFrame(X_resampled, columns=X.columns)
df_Tomek["Class"] = y_resampled

df_Tomek["Class"].value_counts()
```

### Вспомогательные методы для создания моделей

Метод возвращает набор обученных моделей с их названиями

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from keras.models import Sequential
from keras.layers import Dense
import time

def create_models(X_train, y_train):
    models = []
    
    start = time.time()
    logistic_model = LogisticRegression(n_jobs=-1, random_state=42)
    logistic_model.fit(X_train, y_train)
    models.append((logistic_model, "logistic"))
    print(f"logistic - {time.time() - start:.2f}s") # выводим время, сколько обучалась каждая модель
    
    start = time.time()
    forest_model = RandomForestClassifier(n_estimators=100, max_depth=50, n_jobs=-1, random_state=42)
    forest_model.fit(X_train, y_train)
    models.append((forest_model, "random_forest"))
    print(f"random_forest - {time.time() - start:.2f}s")
    
    start = time.time()
    gradient_boosting_model = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=2, random_state=42)
    gradient_boosting_model.fit(X_train, y_train)
    models.append((gradient_boosting_model, "gradient_boosting"))
    print(f"gradient_boosting - {time.time() - start:.2f}s")

    start = time.time()
    model_seq = Sequential()
    model_seq.add(Dense(50, input_dim=X_train.shape[1], activation='relu'))
    model_seq.add(Dense(1, activation='sigmoid'))
    model_seq.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model_seq.fit(X_train, y_train, epochs=50, batch_size=100, validation_split=0.15, verbose=False) # verbose=False
    models.append((model_seq, "sequential"))
    print(f"sequential - {time.time() - start:.2f}s, loss - {model_seq.loss}")
    
    return models
```

Для теста моделей нам понадобится удобный метод визуализации

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_auc_score, roc_curve

def fill_multimatrix(ax, y_test, y_pred_proba, title):
    pred = y_pred_proba[:, -1] > 0.5
    roc_auc = roc_auc_score(y_test, y_pred_proba[:, -1])

    acc = accuracy_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    cm = confusion_matrix(y_test, pred)

    sns.heatmap(cm, annot=True, fmt=".10g", ax=ax)
    ax.set_title(f"{title}\nAccuracy: {acc:.2f}, f1-score: {f1:.2f}, ROC-AUC: {roc_auc:.2f}")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True class")

def draw_multimatrix(models: list, X_test, y_test):
    _, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

    for i in range(len(models)-1):
        pred = models[i][0].predict_proba(X_test)
        ax0 = (i // 2) % 2
        fill_multimatrix(axes[ax0, i % 2], y_test, pred, models[i][1])

    fill_multimatrix(axes[1, 1], y_test, models[-1][0].predict(X_test), models[-1][1])
   
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.model_selection import train_test_split

def test_models(X, y):
    # stratify=y сохраняет пропорцию классов в обучающей и тестирующей выборке
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

    models = create_models(X_train, y_train)
    draw_multimatrix(models, X_test, y_test)
```

### Подбор гиперпараметров для оптимальных моделей

Лучшими наборами данных были наборы, сбалансированные методами SMOTE и ADASYN. Лучшими поделями -- RandomForestClassifier и GradientBoostingClassifier.

Подберём параметры для RandomForestClassifier

```python
from sklearn.model_selection import GridSearchCV

X_train_smote, X_test_smote, y_train_smote, y_test_smote = train_test_split(df_SMOTE.drop(columns=['Class']), df_SMOTE['Class'], test_size=0.2, random_state=42)
X_train_adasyn, X_test_adasyn, y_train_adasyn, y_test_adasyn = train_test_split(df_ADASYN.drop(columns=['Class']), df_ADASYN['Class'], test_size=0.2, random_state=42)

rf_model = RandomForestClassifier()

param_grid = {
    'n_estimators': [75, 100, 125],
    'max_depth': [25, 50, 75],
    'n_jobs': [-1],
    'random_state': [42]
}

# Подбор лучших параметров модели на SMOTE данных
grid_search = GridSearchCV(rf_model, param_grid, cv=3)

grid_search.fit(X_train_smote, y_train_smote)

best_rf_smote_model = RandomForestClassifier(**grid_search.best_params_)
best_rf_smote_model.fit(X_train_smote, y_train_smote)
print(f"Параметры для RandomForestClassifier, SMOTE данные: {grid_search.best_params_}\nScore: {grid_search.best_score_}")

# Подбор лучших параметров модели на ADASYN данных
grid_search.fit(X_train_adasyn, y_train_adasyn)

best_rf_adasyn_model = RandomForestClassifier(**grid_search.best_params_)
best_rf_adasyn_model.fit(X_train_adasyn, y_train_adasyn)
print(f"Параметры для RandomForestClassifier, ADASYN данные: {grid_search.best_params_}\nScore: {grid_search.best_score_}")
```

Подберём параметры для GradientBoostingClassifier

```python
gb_model = GradientBoostingClassifier()

param_grid = {
    'n_estimators': [25, 50, 75],
    'learning_rate': [0.2, 0.9],
    'max_depth': [2, 5],
    'random_state': [42]
}

# Подбор лучших параметров модели на SMOTE данных
grid_search = GridSearchCV(gb_model, param_grid, cv=3, n_jobs=-1)

grid_search.fit(X_train_smote, y_train_smote)

best_gb_smote_model = GradientBoostingClassifier(**grid_search.best_params_)
best_gb_smote_model.fit(X_train_smote, y_train_smote)
print(f"Параметры для GradientBoostingClassifier, SMOTE данные: {grid_search.best_params_}\nScore: {grid_search.best_score_}")

# Подбор лучших параметров модели на ADASYN данных
grid_search.fit(X_train_adasyn, y_train_adasyn)

best_gb_adasyn_model = GradientBoostingClassifier(**grid_search.best_params_)
best_gb_adasyn_model.fit(X_train_adasyn, y_train_adasyn)
print(f"Параметры для RandomForestClassifier, ADASYN данные: {grid_search.best_params_}\nScore: {grid_search.best_score_}")
```

#### Оценим лучшие модели

```python
_, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

pred = best_rf_smote_model.predict_proba(X_test_smote)
fill_multimatrix(axes[0, 0], y_test_smote, pred, "RF, SMOTE")

pred = best_rf_adasyn_model.predict_proba(X_test_adasyn)
fill_multimatrix(axes[0, 1], y_test_adasyn, pred, "RF, ADASYN")

pred = best_gb_smote_model.predict_proba(X_test_smote)
fill_multimatrix(axes[1, 0], y_test_smote, pred, "GB, SMOTE")

pred = best_gb_adasyn_model.predict_proba(X_test_adasyn)
fill_multimatrix(axes[1, 1], y_test_adasyn, pred, "GB, ADASYN")

plt.tight_layout()
plt.show()
```
